# Kernel Density Estimation (KDE) plot

KDE plot is a Kernel Density Estimation Plot which depicts the probability density function of the continuous or non-parametric data variables i.e. we can plot for the univariate or multiple variables altogether.

```python
print(sns.get_dataset_names())
```

```python
#Let's use the 'mpg' dataset
cars = sns.load_dataset('mpg')
print(cars.shape)
cars.head()
```

```python
#Let's make the KDE plot for horsepower
sns.kdeplot(x = 'horsepower', data = cars) # Plot univariate or bivariate distributions using kernel density estimation.
# not a normal distribution
```

```python
sns.kdeplot(x = 'horsepower', data = cars, shade = True)
```

```python
plt.figure(figsize = (15,15)) # change the figure size
sns.kdeplot(x = 'horsepower', data = cars, cumulative = True) # convert to a S-shaped CDF graph
```

```python
sns.kdeplot(x = 'horsepower', data = cars, vertical = True) # reverse the axis
```

```python
sns.kdeplot(x = 'horsepower', data = cars, shade = True, hue = 'origin') # hue used for categorical variables
```

```python
sns.kdeplot(x = 'horsepower', data = cars, color = 'blue', shade = True)
```

```python
# implementation of the Tukey's algorithm
# values that look different from the other values in the data
# Box plots are a visual method to identify outliers
def findOutliers(dataset):
    outliers = []
    Q2 = dataset.median()
    Q1 = dataset.describe()[4]
    Q3 = dataset.describe()[6]
    
    # Q1 = dataset.quantile(0.25)
    # Q3 = dataset.quantile(0.75)
    
    # 
    
    IQR = Q3 - Q1
    lf = Q1 - 1.5*IQR # lf = lower fence
    uf = Q3 + 1.5*IQR # Uf = Upper fence
    for x in dataset:
        if x < lf or x > uf:
            outliers.append(x)
    return outliers
    
```

```python
cars['mpg'].quantile(0.25)
```

```python
findOutliers(cars['mpg'])
```

# Swarmplot: Draw a categorical scatterplot 

```python
myData = [1,1,1,1,1, 2, 2, 2, 3, 4, 4, 4, 4, 4, 5, 5]
sns.swarmplot(x = myData)
```

```python
sns.swarmplot(x = 'horsepower', data = cars);
```

```python
#plt.figure(figsize = (7,7))
sns.swarmplot(x = 'origin',y = 'horsepower', data = cars);
```

```python
sns.stripplot(x = 'origin',y = 'horsepower', data = cars);
```

# Both swarmplot and stripplot are categorical scatter plots used for visualizing data distributions, but they differ in how they handle overlapping points. 
Stipplot: Points may overlap, making it harder to see density.
Swarmplot: Points are spread out horizontally, making density easier to see.

```python
sns.swarmplot(x = 'origin',y = 'horsepower', hue = 'cylinders', data = cars);
```

```python
#Note: Swarmplot does not scale very well when it comes to large amount of data
sns.swarmplot(x = 'origin',y = 'horsepower', hue = 'cylinders', dodge = True, data = cars);
```
